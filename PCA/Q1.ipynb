```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
```

```python
#loading the dataset
x=pd.read_csv('data.csv')
y=pd.read_csv('labels.csv')
x
```

```python
# PCA with dimensionality reduction to 100
# Compute the mean image vector and subtract it from each image vector
x_mean = np.mean(x, axis=0)
X = x - x_mean
X
```

```python
# Compute the covariance matrix of the zero-mean dataset
X_transpose=np.transpose(X)
#using C=1/(n-1)*X.X_tranpose
cov_X = np.dot(X,X_transpose)
cov_X=cov_X/1000
# Compute the eigenvectors and eigenvalues of the covariance matrix
eigvals, eigvector = np.linalg.eig(cov_X)
```

```python
# Sort the eigenvectors in descending order of their corresponding eigenvalues
idx = eigvals.argsort()[::-1]
eigvals = eigenvals[idx]
eigvector = eigenvector[:, idx]
```

```python
# Select the top 100 eigenvectors
eigvectors_100 = eigvector[:, :100]
```

```python
# Project X_tranpose dataset onto the selected  200 eigenvectors
PCA_100 = np.dot(X_transpose, eigvectors_100)
```
